```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the ARuco marker image using OpenCV
aruco_marker_path = "aruco-153.png"
aruco_marker_img = cv2.imread(aruco_marker_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is loaded successfully
if aruco_marker_img is None:
    print("Error: Unable to load the image.")
else:
    # Detect edges in the ARuco marker image
    edges = cv2.Canny(aruco_marker_img, 50, 150)

    # Find contours in the edge-detected image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the bounding box of the ARuco marker
    x, y, w, h = cv2.boundingRect(contours[0])

    # Plot the ARuco marker with bounding box overlay
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.imshow(aruco_marker_img, cmap='gray')
    ax.add_patch(patches.Rectangle((x, y), w, h, edgecolor='r', facecolor='none', linewidth=2))
    ax.set_title('ARuco Marker with Bounding Box')
    ax.axis('off')

    # Show the plot
    plt.show()
```
